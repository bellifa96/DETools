# Questions 
- 1) Récupérer les 3 dates avec le plus de cas confirmés  .
- 2) Récupérer la ligne avec le plus patients guéris.
- 3) Récupérer la somme des décés à EHPAD et à hopital pour chaque jour .

- 5) Récupérer la ligne avec le plus de patients hospitalisés.
- 6) Récupérer la somme des patients hospitalisés et les nouveaux patients hospitalisés .


```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient()
database = client['evaluation']
collection = database['covidfrance']
```

```python
df_espece = pd.read_csv("../data/mongo/covidfrance.csv")
df_espece.head(100)
```

### Netoyer les données


```python
df_espece['date'] = pd.to_datetime(df_espece['date'], errors='coerce')
df_espece = df_espece.dropna(subset=['date'])
 

df_espece.rename(columns={'date':'_id'}, inplace=True)
 
df_espece = df_espece.drop_duplicates(subset='_id', keep='last')
 
l_d =df_espece.to_dict(orient='records')


```

### Importer les données

```python
from pymongo import MongoClient
client = MongoClient("mongo:27017")
bdd_bellifa_espece = client.espece
bellifa_collection_espece = bdd_bellifa_espece['ADBEL-ILLAH']
bellifa_collection_espece.insert_many(l_d)
```

# Question 1

```python
request = bellifa_collection_espece.find().limit(3).sort([("total_cas_confirmes", -1)])
for data in request:
    print(data)
```

# Question 2


```python
request = bellifa_collection_espece.find().limit(1).sort([("total_patients_gueris", -1)])
list(request)



```

# Questions 3

```python
request = bellifa_collection_espece.aggregate(
   [
     {
       '$group':
         {
          '_id': '$_id',
          'somme décés hopital et ehpad': { '$sum': { '$add' : [   '$total_deces_ehpad', '$total_deces_hopital' ]}},
         }
     }
   ]
)
list(request)
```

# Questions 4

```python
request = bellifa_collection_espece.find({"$and":[{'nouveaux_patients_reanimation':{'$gte':500}}]})
list(request)

```

# Questions 5

```python
request = bellifa_collection_espece.find().sort([("patients_hospitalises", -1)])
next(request)
```

# Questions 6

```python
request = bellifa_collection_espece.aggregate(
   [
     {
       '$group':
         {
          '_id': '$_id',
          'patients hospitalisés': { '$sum': { '$add' : [   '$patients_hospitalises', '$nouveaux_patients_hospitalises' ]}},
         }
     }
   ]
)
list(request)
```
